```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt

file_path = '1 - 2019 ODI Cricket Matches.xlsx'
df = pd.read_excel(file_path)

# Extract 'Wins by Runs' information
runs_extracted_runs = df['Margin'].str.extract('(\d+) runs').astype(str)
df['Wins by Runs'] = np.where(runs_extracted_runs == 'nan', 'No Runs', runs_extracted_runs + ' runs')

# Extract 'Wins by Wickets' information
runs_extracted_wickets = df['Margin'].str.extract('(\d+) wickets').astype(str)
df['Wins by Wickets'] = np.where(runs_extracted_wickets == 'nan', 'No Wickets', runs_extracted_wickets + ' wickets')

# Drop the 'Margin' column
df.drop(columns=['Margin'], inplace=True)

# Select relevant columns
df = df[['Team 1', 'Team 2', 'Winner', 'Wins by Runs', 'Wins by Wickets', 'Match Date', 'Scorecard', 'Ground']]

# Count the number of wins for each country
win_counts = df['Winner'].value_counts().head(3)

# Plot the top 3 countries with the most wins
win_counts.plot(kind='bar', figsize=(10, 6))
plt.title('Top 3 Countries with Most Wins')
plt.xlabel('Country')
plt.ylabel('Number of Wins')
plt.show()

# Count the number of matches played at each ground
ground_counts = df['Ground'].value_counts()

# Plotting the top 10 grounds with the most matches
top_grounds = ground_counts.head(10)
top_grounds.plot(kind='barh', color='skyblue', edgecolor='black')
plt.xlabel('Number of Matches')
plt.ylabel('Ground')
plt.title('Top 10 Grounds with the Most Matches')
plt.show()

# Assuming df is your DataFrame with relevant columns
# Filter data related to Pakistan
pak_matches = df[(df['Team 1'] == 'Pakistan') | (df['Team 2'] == 'Pakistan')]

win_pak = pak_matches['Winner'].eq('Pakistan').sum()
loss = pak_matches['Winner'].ne('Pakistan').sum()
# print("Total Matches Played by Pakistan:", len(sri_lanka_matches))
# print("Wins:", pak_matches['Winner'].eq('Pakistan').sum())
# print("Losses:", pak_matches['Winner'].ne('Pakistan').sum())
# plt.bar(['Wins', 'Losses'], [win_pak, loss_pak], color=['green', 'red'])
bars = plt.bar(['Wins', 'Losses'], [win_pak, loss_pak], color=['green', 'red'])
height_of_win = [bar.get_height() for bar in bars][0]
height_of_losses = [bar.get_height() for bar in bars][1]

plt.text('Wins', height_of_win, f'{win_pak}', ha='left', va='bottom', color='black', fontweight='bold')
plt.text('Losses', height_of_losses, f'{loss_pak}', ha='center', va='bottom', color='black', fontweight='bold')

plt.xlabel('Outcome')
plt.ylabel('Number of Matches')
plt.title('Pakistan\'s Performance')
plt.show()
```
